```python
import re

def corregir_espacio(texto):
    texto_corregido = re.sub(r'([a-z])([A-Z])', r'\1 \2', texto)
    return texto_corregido

texto_con_espacio = corregir_espacio("holamellamoCarla")
print(texto_con_espacio)
```

```python
import spacy

def corregir_espacio(texto):
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(texto)
    texto_corregido = " ".join([token.text_with_ws for token in doc])
    return texto_corregido

texto_con_espacio = corregir_espacio("Bonjourje m'appelleCarla")
print(texto_con_espacio)
```

```python
import pandas as pd
import wordninja

# Read the train.csv file
df = pd.read_csv("data/train.csv")

# Function to correct spaces in a news article
def correct_space(news):
    words = wordninja.split(news)
    corrected_news = " ".join(words)
    return corrected_news

# Correct spaces in all news articles
df['corrected_news'] = df['text'].apply(correct_space)

# Save the results to a new file train_spaceChecked.csv
df.to_csv("data/train_spaceChecked.csv", index=False)

# Calculate how many news articles were corrected and their percentage
total_news = len(df)
corrected_news_count = df['corrected_news'].notna().sum()
percentage_corrected = (corrected_news_count / total_news) * 100

print("Number of news articles corrected:", corrected_news_count)
print("Percentage of news articles corrected:", percentage_corrected)
```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize

# Download nltk word tokenizer (only needs to be done once)
nltk.download('punkt')

# Read the train.csv file
df = pd.read_csv("data/train.csv")

# Function to correct spaces in a news article
def correct_space(news):
    words = word_tokenize(news)
    corrected_news = " ".join(words)
    return corrected_news

# Correct spaces in all news articles
df['corrected_news'] = df['text'].apply(correct_space)

# Save the results to a new file train_spaceChecked.csv
df.to_csv("data/train_spaceChecked.csv", index=False)

# Calculate how many news articles were corrected and their percentage
total_news = len(df)
corrected_news_count = df['corrected_news'].notna().sum()
percentage_corrected = (corrected_news_count / total_news) * 100

print("Number of news articles corrected:", corrected_news_count)
print("Percentage of news articles corrected:", percentage_corrected)
```
